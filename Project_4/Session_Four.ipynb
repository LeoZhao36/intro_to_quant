```python
import numpy as np

def bootstrap_mean_ci(data, n_boot=10_000, ci=0.95, seed=None):
    """Bootstrap confidence interval for the mean of a sample.

    data: 1-D array of observations.
    n_boot: number of bootstrap resamples. 10,000 is standard.
    ci: confidence level (0.95 for 95% CI).
    seed: random seed for reproducibility.

    Returns dict with point estimate, CI bounds, and the full
    bootstrap distribution for plotting or further inspection.
    """
    rng = np.random.default_rng(seed)
    data = np.asarray(data)
    n = len(data)

    # Each row of idx is one resample's indices, with replacement.
    idx = rng.integers(0, n, size=(n_boot, n))
    resamples = data[idx]                  # shape (n_boot, n)
    boot_means = resamples.mean(axis=1)    # shape (n_boot,)

    alpha = 1 - ci
    lower = np.quantile(boot_means, alpha / 2)
    upper = np.quantile(boot_means, 1 - alpha / 2)

    return {
        "estimate": data.mean(),
        "ci_lower": lower,
        "ci_upper": upper,
        "boot_distribution": boot_means,
    }
```

```python
import numpy as np

def bootstrap_mean_ci(data, n_boot=10_000, ci=0.95, seed=None):
    """Bootstrap confidence interval for the mean of a sample.

    data: 1-D array of observations.
    n_boot: number of bootstrap resamples. 10,000 is standard.
    ci: confidence level (0.95 for 95% CI).
    seed: random seed for reproducibility.

    Returns dict with point estimate, CI bounds, and the full
    bootstrap distribution for plotting or further inspection.
    """
    rng = np.random.default_rng(seed)
    data = np.asarray(data)
    n = len(data)

    # Each row of idx is one resample's indices, with replacement.
    idx = rng.integers(0, n, size=(n_boot, n))
    resamples = data[idx]                  # shape (n_boot, n)
    boot_means = resamples.mean(axis=1)    # shape (n_boot,)

    alpha = 1 - ci
    lower = np.quantile(boot_means, alpha / 2)
    upper = np.quantile(boot_means, 1 - alpha / 2)

    return {
        "estimate": data.mean(),
        "ci_lower": lower,
        "ci_upper": upper,
        "boot_distribution": boot_means,
    }

def bootstrap_mean_diff_ci(a, b, n_boot=10_000, ci=0.95, seed=None):
    """Bootstrap CI for mean(a) - mean(b).
    Parallel to a two-sample t-test but distribution-free."""
    rng = np.random.default_rng(seed)
    a, b = np.asarray(a), np.asarray(b)
    na, nb = len(a), len(b)

    idx_a = rng.integers(0, na, size=(n_boot, na))
    idx_b = rng.integers(0, nb, size=(n_boot, nb))

    diffs = a[idx_a].mean(axis=1) - b[idx_b].mean(axis=1)

    alpha = 1 - ci
    return {
        "estimate": a.mean() - b.mean(),
        "ci_lower": np.quantile(diffs, alpha / 2),
        "ci_upper": np.quantile(diffs, 1 - alpha / 2),
        "boot_distribution": diffs,
    }
```
