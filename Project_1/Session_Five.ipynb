```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the four series saved from Session 4
series_files = {
    'HS300 full':      'data/basket_returns_hs300_full.csv',
    'HS300 ex-stim':   'data/basket_returns_hs300_exstim.csv',
    'ZZ1000 full':     'data/basket_returns_zz1000_full.csv',
    'ZZ1000 ex-stim':  'data/basket_returns_zz1000_exstim.csv',
}

series = {}
for name, path in series_files.items():
    s = pd.read_csv(path, index_col=0, parse_dates=True).squeeze().dropna()
    series[name] = s
    print(f"{name}: N = {len(s)}, mean = {s.mean():.5f}, std = {s.std():.5f}")
```

```python
results = []
for name, s in series.items():
    jb_stat, jb_p = stats.jarque_bera(s)
    sw_stat, sw_p = stats.shapiro(s)
    results.append({
        'Series':    name,
        'N':         len(s),
        'Skew':      round(s.skew(), 3),
        'ExKurt':    round(s.kurtosis(), 3),
        'JB_stat':   round(jb_stat, 1),
        'JB_p':      jb_p,
        'SW_W':      round(sw_stat, 4),
        'SW_p':      sw_p,
    })

results_df = pd.DataFrame(results)

# Format p-values in scientific notation so you can see the exponents
with pd.option_context('display.float_format', '{:.6g}'.format):
    print(results_df.to_string(index=False))
```
