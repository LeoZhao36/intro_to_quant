```python
import baostock as bs 
import pandas as pd 

lg = bs.login()
print(lg.error_msg)

# Pull daily K-line data for 平安银行, last 1 year, with 前复权 (forward-adjusted prices)
rs = bs.query_history_k_data_plus(
    "sz.000001",
    "date,open,high,low,close,volume",
    start_date="2024-04-01",
    end_date="2025-04-01",
    frequency="d",       # d = daily
    adjustflag="2"        # 2 = 前复权 (forward adjustment)
)

# Convert the result into a pandas DataFrame
data_list = []
while (rs.error_code == '0') and rs.next():
    data_list.append(rs.get_row_data())

df = pd.DataFrame(data_list, columns=rs.fields)
```

```python
# Convert price columns from strings to floats
for col in ['open', 'high', 'low', 'close', 'volume']:
    df[col] = pd.to_numeric(df[col])

# Convert date column from string to datetime
df['date'] = pd.to_datetime(df['date'])

# Set date as the index (the row label)
df = df.set_index('date')
```

```python
# Subtract the entire 'low' column from the entire 'high' column, all at once
df['range'] = df['high'] - df['low']

# Show the result
df[['high', 'low', 'range']].head()

# How many rows have a range value?
df['range'].count()

# Find the maximum value
df['range'].max()

# Find WHICH day had the maximum
df['range'].idxmax()
```
